# Interactive Visualization Dashboard

```python
# Import libraries and modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
from scr.visualization import (
    plot_skill_frequency,
    plot_skills_by_title_heatmap,
    plot_title_distribution,
    plot_skill_associations,
    plot_correlation_heatmap,
    plot_skill_network,
    create_interactive_skill_plot,
    create_comprehensive_dashboard
)
from scr.skill_analysis import analyze_skill_frequency, analyze_skills_by_title, get_title_statistics
from scr.correlation_analysis import create_skill_cooccurrence_matrix, analyze_skill_associations, calculate_statistical_correlations
```

```python
# Load analyzed data
df = pd.read_csv('data/processed/cleaned_job_data.csv')
print(f"Loaded data shape: {df.shape}")
```

## Generate analysis data for visualization

```python
print("Generating analysis data for visualizations...")
skill_freq = analyze_skill_frequency(df, 'skills')
skills_by_title = analyze_skills_by_title(df, 'title_category', 'skills')
title_stats = get_title_statistics(df, 'title_category', 'skills')
cooccurrence_matrix = create_skill_cooccurrence_matrix(df, 'skills')
python_associations = analyze_skill_associations(df, 'python', 'skills')
corr_matrix, _ = calculate_statistical_correlations(df, 'skills')
```

## Create individual visualizations

```python
print("\nCreating individual visualizations...")

# 1. Skill frequency plot
plot_skill_frequency(skill_freq, top_n=15, 
                    title="Top Skills Across All Job Postings")
```

```python
# 2. Skills by title heatmap
plot_skills_by_title_heatmap(skills_by_title, top_n_titles=8, top_n_skills=12,
                            normalize=True)
```

```python
# 3. Title distribution
plot_title_distribution(title_stats, top_n=10)
```

```python
# 4. Python associations
plot_skill_associations(python_associations, 'python', top_n=15)
```

```python
# 5. Correlation heatmap
plot_correlation_heatmap(corr_matrix, top_n_skills=15)
```

```python
# 6. Skill network (if networkx is available)
try:
    plot_skill_network(cooccurrence_matrix, min_cooccurrence=3)
except ImportError:
    print("NetworkX not available for network visualization")
```

```python
# 7. Interactive plot
create_interactive_skill_plot(skill_freq, top_n=20)
```

```python
# Alternatively, create comprehensive dashboard
print("\nCreating comprehensive dashboard...")
create_comprehensive_dashboard(
    df, 
    title_column='title_category', 
    skills_column='skills',
    save_dir='outputs/figures/'
)
```

```python
# Create detailed analysis for specific titles
print("\nDetailed analysis for specific titles:")
target_titles = ['Senior Data Scientist', 'Senior Data Analyst', 'Senior Machine Learning Engineer']

for title in target_titles:
    print(f"\n=== {title} ===")
    title_df = df[df['title_category'] == title]
    if len(title_df) > 0:
        title_skills = analyze_skill_frequency(title_df, 'skills')
        print(f"Top skills for {title}:")
        print(title_skills.head(8))
    else:
        print(f"No data found for {title}")

print("\nVisualization dashboard complete!")
print("All figures saved to outputs/figures/")
```

## 4. Analyze Skill Frequency Results
Examine the most in-demand skills in the data science job market.

```python
# Analyze skill frequency results
skill_freq = report['skill_frequency']

print("üîß SKILL FREQUENCY ANALYSIS")
print("=" * 50)

print(f"\nüìä Overall Statistics:")
print(f"   Total skills mentioned: {skill_freq['total_skills_mentioned']:,}")
print(f"   Unique skills: {skill_freq['unique_skills']}")
print(f"   Average skills per job: {skill_freq['avg_skills_per_job']:.1f}")

print(f"\nüèÜ Top 10 Most Demanded Skills:")
top_skills = skill_freq['top_skills'][:10]
for i, (skill, count) in enumerate(top_skills, 1):
    prevalence = skill_freq['skill_prevalence'][skill]
    print(f"   {i:2d}. {skill:25} {count:4d} mentions ({prevalence:.1f}% of jobs)")
```

```python
# Visualize skill frequency distribution
plt.figure(figsize=(12, 8))
skills, counts = zip(*skill_freq['top_skills'][:15])

plt.subplot(1, 2, 1)
plt.barh(skills, counts, color=plt.cm.Set3(np.arange(len(skills))))
plt.xlabel('Number of Mentions')
plt.title('Top 15 Most Mentioned Skills')
plt.gca().invert_yaxis()

plt.subplot(1, 2, 2)
# Skill prevalence
prevalence_data = [(skill, skill_freq['skill_prevalence'][skill]) for skill in skills]
skills, prevalence = zip(*prevalence_data)
plt.barh(skills, prevalence, color=plt.cm.Set2(np.arange(len(skills))))
plt.xlabel('Percentage of Jobs (%)')
plt.title('Skill Prevalence in Job Postings')
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

## 5. Analyze Skill Categories
Examine the distribution of skills across different categories.

```python
# Analyze skill categories
category_analysis = report['skill_categories']

print("üìö SKILL CATEGORY ANALYSIS")
print("=" * 50)

print(f"\nüìä Skill Category Prevalence:")
categories_sorted = sorted(category_analysis.items(), 
                          key=lambda x: x[1]['percentage'], reverse=True)

for category, data in categories_sorted:
    print(f"   {category:20} {data['percentage']:5.1f}% of jobs ({data['count']} jobs)")

# Identify most and least common categories
most_common = categories_sorted[0]
least_common = categories_sorted[-1]

print(f"\nüéØ Key Insights:")
print(f"   Most common category: {most_common[0]} ({most_common[1]['percentage']:.1f}%)")
print(f"   Least common category: {least_common[0]} ({least_common[1]['percentage']:.1f}%)")
```

## 6. Role-Based Skill Comparison
Compare skill requirements between different data science roles.

```python
# Check if we have role categories for comparison
if 'role_category' in df.columns and df['role_category'].nunique() >= 2:
    print("üéØ ROLE-BASED SKILL COMPARISON")
    print("=" * 50)
    
    # Get top two role categories for comparison
    top_roles = df['role_category'].value_counts().head(2).index.tolist()
    role1, role2 = top_roles[0], 'Data Analyst'
    
    print(f"\nComparing: {role1} vs {role2}")
    
    # Use the analyzer method for role comparison
    try:
        correlation_analysis, title1, title2 = analyzer.analyze_skills_correlation_between_titles(df, role1, role2)
        
        print(f"\nüìä Skill Requirements Comparison:")
        for category, data in correlation_analysis.items():
            diff = data['difference']
            diff_symbol = "+" if diff > 0 else ""
            print(f"   {category:20} {role1}: {data['title_1_jobs_percentage']:5.1f}% | "
                  f"{role2}: {data['title_2_jobs_percentage']:5.1f}% | "
                  f"Diff: {diff_symbol}{diff:+.1f}%")
        
        # Generate comparison plot
        analyzer.plot_ai_skill_correlations(correlation_analysis)
        
    except Exception as e:
        print(f"‚ùå Role comparison failed: {e}")
else:
    print("‚ö†Ô∏è  Insufficient role categories for comparison")
```

## 7. Analyze Skills by Seniority Level
Examine how skill requirements vary by experience level.

```python
# Analyze skills by seniority
seniority_analysis = report['seniority_analysis']

if seniority_analysis:
    print("üë®‚Äçüíº SENIORITY-BASED SKILL ANALYSIS")
    print("=" * 50)
    
    for level, data in seniority_analysis.items():
        print(f"\nüìã {level.upper()} LEVEL:")
        print(f"   Number of jobs: {data['job_count']}")
        print(f"   Average skills per job: {data['avg_skills_per_job']:.1f}")
        print(f"   Top 5 skills:")
        
        for i, (skill, count) in enumerate(data['top_skills'][:5], 1):
            print(f"      {i}. {skill} ({count} mentions)")
else:
    print("‚ö†Ô∏è  Seniority data not available for analysis")
```

## 8. Geographic Analysis
Examine how skill requirements vary by location.

```python
# Analyze skills by location
location_analysis = report['location_analysis']

if location_analysis:
    print("üåç GEOGRAPHIC SKILL ANALYSIS")
    print("=" * 50)
    
    for location, data in location_analysis.items():
        if data['job_count'] > 10:  # Only show locations with sufficient data
            print(f"\nüìç {location.upper()}:")
            print(f"   Number of jobs: {data['job_count']}")
            print(f"   Average skills per job: {data['avg_skills_per_job']:.1f}")
            print(f"   Most demanded skills:")
            
            for i, (skill, count) in enumerate(data['top_skills'][:3], 1):
                print(f"      {i}. {skill}")
else:
    print("‚ö†Ô∏è  Location data not available for analysis")
```

## 9. Temporal Trends Analysis
Examine how skill requirements have evolved over time.

```python
# Analyze temporal trends
temporal_analysis = report['temporal_trends']

if temporal_analysis:
    print("üìÖ TEMPORAL SKILL TRENDS")
    print("=" * 50)
    
    # Get most recent and oldest periods
    periods = sorted(temporal_analysis.keys())
    if len(periods) >= 2:
        recent_period = periods[-1]
        old_period = periods[0]
        
        print(f"\nüïê Comparing {old_period} vs {recent_period}:")
        
        recent_skills = [skill for skill, _ in temporal_analysis[recent_period]['top_skills']]
        old_skills = [skill for skill, _ in temporal_analysis[old_period]['top_skills']]
        
        print(f"   Recent top skills: {', '.join(recent_skills)}")
        print(f"   Older top skills: {', '.join(old_skills)}")
        
        # Identify emerging skills (in recent but not in old)
        emerging_skills = set(recent_skills) - set(old_skills)
        if emerging_skills:
            print(f"   üî• Emerging skills: {', '.join(emerging_skills)}")
        
        # Identify declining skills (in old but not in recent)
        declining_skills = set(old_skills) - set(recent_skills)
        if declining_skills:
            print(f"   üìâ Declining skills: {', '.join(declining_skills)}")
else:
    print("‚ö†Ô∏è  Temporal data not available for analysis")
```

## 10. Skill Combinations Analysis
Examine which skills are most frequently required together.

```python
# Analyze skill combinations
combo_analysis = report['skill_combinations']

print("üîÑ SKILL COMBINATIONS ANALYSIS")
print("=" * 50)

print(f"\nüìä Combination Statistics:")
print(f"   Total skill pairs: {combo_analysis['total_combinations']:,}")
print(f"   Unique combinations: {combo_analysis['unique_combinations']}")

print(f"\nü§ù Top 10 Most Common Skill Pairs:")
top_combos = combo_analysis['top_combinations'][:10]
for i, ((skill1, skill2), count) in enumerate(top_combos, 1):
    print(f"   {i:2d}. {skill1:20} + {skill2:20} ({count} occurrences)")
```

## 11. Advanced Insights and Correlations
Generate deeper insights from the analysis results.

```python
print("üí° ADVANCED INSIGHTS")
print("=" * 50)

# Insight 1: Most versatile skills (appear in most categories)
print(f"\nüéØ Key Market Insights:")

# Identify high-prevalence skills
high_prevalence_skills = [(skill, prev) for skill, prev in skill_freq['skill_prevalence'].items() 
                          if prev > 50]  # Skills required in >50% of jobs

if high_prevalence_skills:
    print(f"1. Core Skills (required in >50% of jobs):")
    for skill, prev in sorted(high_prevalence_skills, key=lambda x: x[1], reverse=True)[:5]:
        print(f"   ‚úì {skill} ({prev:.1f}% of jobs)")

# Insight 2: Regional variations
if location_analysis:
    locations_with_data = [(loc, data) for loc, data in location_analysis.items() if data['job_count'] > 10]
    if len(locations_with_data) >= 2:
        # Compare top skills across locations
        location_skills = {}
        for location, data in locations_with_data:
            top_skill = data['top_skills'][0][0] if data['top_skills'] else "Unknown"
            location_skills[location] = top_skill
        
        print(f"2. Regional Specializations:")
        for location, top_skill in location_skills.items():
            print(f"   üìç {location}: {top_skill}")

# Insight 3: Skill density by role type
if 'role_category' in df.columns:
    role_skill_density = df.groupby('role_category')['skills_count'].mean().sort_values(ascending=False)
    if len(role_skill_density) > 0:
        most_dense_role = role_skill_density.index[0]
        least_dense_role = role_skill_density.index[-1]
        print(f"3. Skill Density by Role:")
        print(f"   üìä Most skill-intensive: {most_dense_role} ({role_skill_density[most_dense_role]:.1f} skills/job)")
        print(f"   üìä Least skill-intensive: {least_dense_role} ({role_skill_density[least_dense_role]:.1f} skills/job)")
```

Save the analysis findings for reporting and future reference.
